```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier 
from sklearn import naive_bayes as nb
```

```python
# Import the datasets
data_train = pd.read_csv('.\datasets/train.csv', header=None)
data_test = pd.read_csv('.\datasets/test.csv', header=None)

# Store the unprocessed features in arrays
ini_train = np.array(data_train.iloc[1:,:11])
ini_test = np.array(data_test.iloc[1:,:11])

# Store classes in arrays
y_train = np.array(data_train.iloc[1:,11])
y_test = np.array(data_test.iloc[1:,11])

# Take a look at the data
data_train
```

```python
# Preprocessing
# TO DO?
# Check if any data is negative (it shouldn't be)
# (Should merge data and resplit with validation set?)
# Weight data?

# This just transforms the data into standardized vectors
def preprocess(ar):
    new_data = np.zeros(11)
    new_data = np.reshape(new_data, (1, 11))
    
    for line in ar:
        new_line = np.array([])
        
        # I know this is like. not efficient but I don't know how to code Python
        new_line = np.append(new_line, int(line[0]))
        new_line = np.append(new_line, float(line[1]))
        new_line = np.append(new_line, int(line[2]))
        new_line = np.append(new_line, float(line[3]))
        new_line = np.append(new_line, int(line[4]))
        new_line = np.append(new_line, int(line[5]))
        new_line = np.append(new_line, int(line[6]))
        new_line = np.append(new_line, int(line[7]))
        new_line = np.append(new_line, int(line[8]))
        new_line = np.append(new_line, int(line[9]))
        new_line = np.append(new_line, int(line[10]))
        
        new_line = np.reshape(new_line, (1, 11))
        new_data = np.append(new_data, new_line, axis=0)
    
    new_data = new_data[1:]
    
    scaler = StandardScaler()
    new_data = scaler.fit_transform(new_data)
            
    return new_data
```

```python
# Preprocess the training and testing datasets
X_train = preprocess(ini_train)
X_test = preprocess(ini_test)
```

```python
# SVM
svm_clf = svm.LinearSVC(max_iter=1500)
svm_clf.fit(X_train, y_train)

print(svm_clf.score(X_train, y_train))
print(svm_clf.score(X_test, y_test)) 
```

```python
# RFC
rfc = RandomForestClassifier(n_estimators=100, max_depth=2)
rfc.fit(X_train, y_train)

print(rfc.score(X_train, y_train)) # This is overfitted
print(rfc.score(X_test, y_test))
```

```python
# Naive Bayes
nbc = nb.BernoulliNB()
nbc.fit(X_train, y_train)

print(nbc.score(X_train, y_train))
print(nbc.score(X_test, y_test))
```
